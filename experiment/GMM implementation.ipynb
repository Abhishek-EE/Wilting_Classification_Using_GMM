```python
def load_data_train():
    train_data_path = Data_Path + 'TrainData//'
    dfData = pd.read_csv(Data_Path + 'TrainAnnotations.csv')
    #df = dfData.loc[dfData['annotation']==i]    
    df = image_to_feature(dfData,train_data_path)
    np_df = PCA_to_reqdFeaturesoriginal(df)
    np_df['labels'] = dfData.annotation
    
    return np_df
```

```python
def image_to_feature(df,path):
    'takes a df and converts it to images'
    f = []
    start = time.time()
    for i in df.file_name:
        image = cv2.imread(path+i)
        #print(type(image))
        dimg = feature.feature_extraction(img = image)
        f.append(dimg)
    imgarray = np.zeros((len(f),f[0].shape[1]))
    for i in range(len(f)):
        imgarray[i,:] = f[i]
    df = pd.DataFrame(imgarray)
    return df
```

```python
def PCA_to_reqdFeaturesoriginal(df):
    pca = PCA(0.9)
    np_ = pca.fit_transform(df)
    df_ = pd.DataFrame(np_)
#     train_set = df_copy.sample(frac=0.75, random_state=0)
#     val_set = df_copy.drop(train_set.index)
    return df_
    
    
```

```python
def PCA_to_reqdFeatures(data):
    df_copy = data.copy()
    train_set = df_copy.sample(frac=0.75, random_state=0, axis = 0)
    val_set = df_copy.drop(train_set.index)
    return train_set , val_set

    
```
